Función para calcular la nota de la cursada. A esta nota se le suma luego los Python plus y un punto más si hacen el informe final. 

```python
def nota_base(puntos_cursada): 
    """ Calcula la nota de la cursada """
    print(puntos_cursada)
    if puntos_cursada >= 70 and puntos_cursada < 77.5:
        return 4
    elif puntos_cursada  >= 77.5 and puntos_cursada < 85:
        return 5
    elif puntos_cursada  >= 85 and puntos_cursada < 92.5:
        return 6
    elif puntos_cursada  >= 92.5 and puntos_cursada <=100:
        return 7
    else:
        return 0
```
